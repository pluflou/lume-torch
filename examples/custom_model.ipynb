# Creating a Custom LUME-model
Custom models that are compatible with LUME tools can be created by inhereting from the `LUMETorch`.

```python
from lume_torch.base import LUMETorch
from lume_torch.variables import TorchScalarVariable
```

## Model Definition
The minimum requirement for creating a custom LUME-model is to implement the abstract `evaluate` method inherited from `LUMETorch`. Here, we simply return the squared input.

```python
class ExampleModel(LUMETorch):
    def _evaluate(self, input_dict):
        output_dict = {
            "output1": input_dict[self.input_variables[0].name] ** 2,
            "output2": input_dict[self.input_variables[1].name] ** 2,
        }
        return output_dict
```

## Model Instantiation and Execution
Instantiation requires specification of the input and output variables of the model.

```python
input_variables = [
    TorchScalarVariable(name="input1", default_value=0.1),
    TorchScalarVariable(name="input2", default_value=0.2, value_range=[0.0, 1.0]),
]
output_variables = [
    TorchScalarVariable(name="output1"),
    TorchScalarVariable(name="output2"),
]

m = ExampleModel(input_variables=input_variables, output_variables=output_variables)
```

```python
input_dict = {
    "input1": 0.3,
    "input2": 0.6,
}
m.evaluate(input_dict)
```

# Save to YAML

```python
m.dump("example_model.yml")
```

# Load from YAML

```python
m = ExampleModel("example_model.yml")
```
